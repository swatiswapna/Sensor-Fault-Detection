```python
# we will write two methods where we can go for check std and null values 
```

```python
def check_std(df : pd.DataFrame):
    drop_col = []
    num_cols = [col for col in df.columns if df[col].dtype != 'O']
    for col in num_cols:
        if df[col].std() == 0:
            drop_col.append(col)
    return drop_col
```

```python
def check_null(df: pd.DataFrame , thresold = .7):
    cols_missing_ratios = df.isna().sum().div(df.shape[0])
    cols_to_drop = list(cols_missing_ratios[cols_missing_ratios > thresold].index)
    return cols_to_drop 
```

```python
cols_drop_null = check_null(df1,thresold=.7)
cols_drop_null
```

```python
cols_drop_std = check_std(df1)
cols_drop_std
```

```python
drop_cols  = cols_drop_std + cols_drop_null
```

```python
X, y = df1.drop(drop_cols, axis=1), df[["Good/Bad"]]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import KNNImputer    # impute nearest mean points 
from sklearn.impute import SimpleImputer  # MEAN IMPUTE 
from sklearn.preprocessing import RobustScaler
```

```python
imputer = KNNImputer(n_neighbors=3) 
preprocessing_pipeline = Pipeline(
    steps=[('imputer',imputer) , ('scaler',RobustScaler())]
)
```

```python
preprocessing_pipeline
```

```python
X_transform = preprocessing_pipeline.fit_transform(X)
```

```python
X_transform.shape
```

```python
X_transform
```

Time to do some data Sampling to make it balance from imbalance 

```python
plt.scatter(X_transform[0],X_transform[1], label='Data Points')
```

```python
from imblearn.combine import SMOTETomek
```

```python
resampler = SMOTETomek(sampling_strategy='auto')

X_resample , y_resample = resampler.fit_resample(X_transform,y)
```

Data is ready to train the model 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train, y_test  = train_test_split(X_resample,y_resample,test_size=0.2,random_state=42)
```

```python
print(f"train set: {X_train.shape, y_train.shape}")
print(f"test set: {X_test.shape, y_test.shape}")
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_auc_score
```

```python
model_sigmoid_svc = SVC(kernel='sigmoid',verbose=2)
model_rbf_svc = SVC(kernel='rbf',verbose=2)
model_randomforest = RandomForestClassifier(verbose=2)
model_xgb= XGBClassifier()
```

```python
#model_sigmoid_svc

score = cross_val_score(model_sigmoid_svc,X_train,y_train,scoring='roc_auc',cv = 10 , verbose=2)
```

```python
y_pred = cross_val_predict(model_sigmoid_svc,X_test,y_test,cv=5)
svc_auc = roc_auc_score(y_test, y_pred)
svc_auc
```

```python
score = cross_val_score(model_rbf_svc,X_train,y_train,scoring='roc_auc',cv = 10 , verbose=2)
```

```python
y_pred = cross_val_predict(model_rbf_svc,X_test,y_test,cv=5)
svc_auc = roc_auc_score(y_test, y_pred)
svc_auc
```

```python
score = cross_val_score(model_randomforest,X_train,y_train,scoring='roc_auc',cv = 10 , verbose=2)
```
